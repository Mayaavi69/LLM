```python
# 1.1
import torch
from transformers import pipeline
```

```python
# 1.2 Initial chat

chat = [
        {"role": "system", "content": "You are a sassy, wise-cracking robot as imagined by Hollywood circa 1986."},
        {"role": "user", "content": "Hey, can you tell me any fun things to do in New York?"}
       ]
```

```python
%%time

# 1.3 Pipeline to get reply:

pipe = pipeline("text-generation",
                "TinyLlama/TinyLlama-1.1B-Chat-v1.0",
                torch_dtype=torch.bfloat16,
                device_map="auto"
                )

```

```python
%%time

# 2.0 Get response to our chat:

response = pipe(chat,
                max_new_tokens=512
                )

# 2.1 Print messages:
print(response[0]['generated_text'][-1]['content'])
```

```python
# 2.2
chat = response[0]['generated_text']

# 2.2.1 Append current question to earlier chat:

chat.append(
            {"role": "user", "content": "Tell me more about the museum at point 9"}
           )

```

```python
%%time

# 3.0 Get response:

response1 = pipe(chat, max_new_tokens=512)
print(response1[0]['generated_text'][-1]['content'])
```

```python
# 4.0
chat = response1[0]['generated_text']

# 4.1 Append current question to earlier chat:

chat.append(
            {"role": "user", "content": "Can you tell what all got destroyed in the fire that you mentioned in point 1"}
           )

```

```python
%%time

# 4.2 Get response:

response2 = pipe(chat, max_new_tokens=1000)
print(response2[0]['generated_text'][-1]['content'])
```
