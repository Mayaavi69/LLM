```python
from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent
import pandas as pd
# https://python.langchain.com/v0.1/docs/integrations/toolkits/pandas/
# https://stackoverflow.com/a/77401735/3282777
```

```python
# 1.0
from langchain_community.llms import Ollama

# 1.0.1
llm = Ollama(model="codellama:13b")
```

```python
llm.invoke("Are you aware of pandas code?")
```

```python
df = pd.read_csv(
    "https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv"
)
df.head()
```

```python
agent = create_pandas_dataframe_agent(llm,
                                      df,
                                      verbose=True,
                                      agent_type="zero-shot-react-description",
                                      return_intermediate_steps=True
                                      )
```

```python
agent.invoke("In dataframe, df, find correlation between Age and Fare columns")
```

```python
agent.invoke("count the number of  people having more than two siblings")
```
